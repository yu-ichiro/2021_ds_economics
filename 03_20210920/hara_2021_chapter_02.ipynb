```python
#sklearn からdataset を利用する
from sklearn import datasets
#pandas パッケージを利用する
import pandas as pd
#load_boston でデータセットのサンプルを読み込む
dataset = datasets.load_boston()
#DataFrame に代入する
dataset2=pd.DataFrame(dataset.data)
#列名を指定する
dataset2.columns=dataset.feature_names
#価格を取り出す
dataset3=pd.DataFrame(dataset.target)
#列名を任意に指定する
dataset3.columns=['PRICE']
#ふたつのデータセットを縦につなぎ合わせる
dataset4=pd.concat([dataset3, dataset2], axis=1)
#結果を出力する
dataset4
```

```python
#単回帰分析を実行してみる
#価格に対してNOXがマイナスに寄与することが確認できる

import statsmodels.api as sm
model = sm.OLS(dataset4.PRICE, sm.add_constant(dataset4.NOX))
results = model.fit()
print(results.summary())
print('p-values\n', results.pvalues)
```
