楽天市場のデータセットを取得する

```python
import json

with open("../credential.json") as fp:
    credential = json.load(fp)
```

```python
import urllib.parse
import urllib.error

keyword="うさまる"
applicationId=credential["rakuten"]["applicationId"]
url=f"https://app.rakuten.co.jp/services/api/IchibaItem/Search/20170706" \
    f"?format=json" \
    f"&keyword={urllib.parse.quote(keyword)}" \
    f"&applicationId={applicationId}"
```

```python
import urllib.request as request

contents = ""
try:
    response = request.urlopen(url)
    contents = response.read().decode()
    print(contents)
    response.close()
except urllib.error.HTTPError:
        print("not available")
else:
    pass
```

```python
import pandas as pd
import json
temp2=json.loads(contents)
temp3=pd.json_normalize(temp2['Items'])
```

楽天トラベル

```python
# 検索する位置情報の設定
latitude="128440.51"
longitude="503172.21"
searchRadius="1"
url = f"https://app.rakuten.co.jp/services/api/Travel/SimpleHotelSearch/20170426" \
      f"?format=json" \
      f"&applicationId={applicationId}" \
      f"&latitude={latitude}" \
      f"&longitude={longitude}" \
      f"&searchRadius={searchRadius}"
```

```python
url
```

```python
try:
    response = request.urlopen(url)
    contents = response.read().decode()
    print(contents)
    response.close()
except urllib.error.HTTPError:
        print("not available")
else:
    pass
```

```python
import pandas as pd
import json
temp2=json.loads(contents)
temp3=pd.json_normalize(temp2['hotels'], record_path='hotel')
```
